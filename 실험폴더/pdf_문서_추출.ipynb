```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# 허깅페이스 Llama-3-Open-Ko-8B-Instruct-preview 모델 로드
model_name = "beomi/Llama-3-Open-Ko-8B-Instruct-preview"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# GPU 사용 가능 시 모델을 GPU로 이동
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

def generate_response(input_text):
    # 입력 텍스트를 토큰화 및 텐서로 변환
    inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)
    inputs = {key: value.to(device) for key, value in inputs.items()}
    
    # 모델에 입력하여 응답 생성
    outputs = model.generate(**inputs, max_length=1024, num_return_sequences=1)
    
    # 생성된 응답을 디코딩하여 텍스트로 변환
    response_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    return response_text

```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
from tqdm import tqdm

# 허깅페이스 Llama-3-Open-Ko-8B-Instruct-preview 모델 로드
model_name = "beomi/Llama-3-Open-Ko-8B-Instruct-preview"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# GPU 사용 가능 시 모델을 GPU로 이동
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

def generate_response(input_text):
    # 입력 텍스트를 토큰화 및 텐서로 변환
    inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)
    inputs = {key: value.to(device) for key, value in inputs.items()}
    
    # 모델에 입력하여 응답 생성
    # tqdm을 사용하여 진행 바 표시
    with tqdm(total=100, desc="Generating Response") as pbar:
        outputs = model.generate(**inputs, max_length=1024, num_return_sequences=1)
        pbar.update(100)  # 진행 바 업데이트
    
    # 생성된 응답을 디코딩하여 텍스트로 변환
    response_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    return response_text

print("챗봇을 시작합니다. 종료하려면 '종료'라고 입력하세요.")
while True:
    user_input = input("User: ")
    if user_input.lower() in ["종료", "exit", "quit"]:
        print("챗봇을 종료합니다.")
        break
    
    # 사용자 입력에 대한 응답 생성
    response = generate_response(user_input)
    print(f"Bot: {response}")
```
